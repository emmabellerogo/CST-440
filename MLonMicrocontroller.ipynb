# CLC - Machine Learning on a Microcontroller #
## By: Andru Tjalas, Josh Nelson, Martin Battu, and Emma Rogoveanu ##

### 1. Introduction & Objective ###

The objective of this project is to design, train, and deploy a machine learning model to approximate trigonometric functions, specifically sin(x), cos(x), and tan(x). The goal of this project is to demonstrate that a neural network can learn these functions and be integrated into an embedded workflow using an Arduino Nano, with output captured.
